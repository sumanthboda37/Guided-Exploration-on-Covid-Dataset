```python
#7.5 points
#Task 1-c: Print the total cases confirmed, deaths, and recovered cases of each country with the given dataset.
#The below variable holds data for all the unique countries, and their total - confirmed, deaths, and recovered cases.
df1_data = pd.read_csv('covid.csv',encoding = 'latin1',usecols = ['Country/Region','Confirmed','Deaths','Recovered'])
total_confirmed_deaths_recovered = df1_data.groupby(["Country/Region"]).max()
print ("\n\n>>Task 1-c: Total cases confirmed, deaths, and recovered are \n", total_confirmed_deaths_recovered)
```

```python
# 5 points for both questions below,

#Task 1-d-i: Print top 10 worst affected countries with confirmed cases.
top_confirmed_ten_countries  = df1_data.groupby(["Country/Region"]).max()
top_confirmed_ten_countries = top_confirmed_ten_countries.sort_values(by = "Confirmed",ascending = False)
print ("\n\n >>Task 1-d: top 10 worst affected Countries by Confirmed :\n", top_confirmed_ten_countries.head(10))
#Task 1-d-ii: Print top 5 worst affected countries with death cases.
top_death_five_countries  = df1_data.groupby(["Country/Region"]).max()
top_death_five_countries = top_death_five_countries.sort_values(by = "Deaths",ascending = False)
print ("\n\n >>Task 1-d: top 5 worst affected countries by Death\n", top_death_five_countries.head(5))
```

# Task 2: Aggregation & Filtering & Rank
In this task, we will perform some very high level aggregation and filtering operations. 
Then, we will apply ranking on the results for some tasks. 
Pandas has a convenient and powerful syntax for aggregation, filtering, and ranking. 
DO NOT write a for loop. Pandas has built-in functions for all tasks. 

```python
# 8 points
#Task 2-a: Find out the countries that has had more than a total of 2 hundred thousand confirmed cases.
confirmed_greater_than_200K = df1_data.groupby(["Country/Region"]).max()
confirmed_greater_than_200K = confirmed_greater_than_200K[(confirmed_greater_than_200K ['Confirmed']>200000.0)]
print (">>Task 2-a:  The countries that has had more than a total of 2 hundred thousand confirmed cases are: \n%s" % (confirmed_greater_than_200K))

# 8 points
#Task 2-b: Find out the total number of confirmed, recovered, and death cases for each WHO region.
#df2_data = pd.read_csv('covid.csv',encoding = 'latin1',usecols = ['WHO Region','Confirmed','Deaths','Recovered'])
total_cases_whoregion = df_data.groupby(["Country/Region"]).max()
total_cases_whoregion1 = total_cases_whoregion.groupby(["WHO Region"]).aggregate({'Confirmed':'sum','Deaths':'sum','Recovered':'sum'})
print ("\n\n>>Task 2-b: The total number of confirmed, recovered, and death cases for each WHO region are: \n%s" % (total_cases_whoregion1))

# 7 points
#Task 2-c: Find out the top 5 poorly performing countries, in the order of confirmed, deaths, and recovered cases.
df3_data = pd.read_csv('covid.csv',encoding = 'latin1',usecols = ['Country/Region','Confirmed'])
top5_poorlyperforming_confirmed= df3_data.groupby(["Country/Region"]).max()
top5_poorlyperforming_confirmed= top5_poorlyperforming_confirmed.sort_values(by = "Confirmed",ascending = False)
df4_data = pd.read_csv('covid.csv',encoding = 'latin1',usecols = ['Country/Region','Deaths'])
top5_poorlyperforming_deaths = df4_data.groupby(["Country/Region"]).max()
top5_poorlyperforming_deaths = top5_poorlyperforming_deaths.sort_values(by = "Deaths",ascending = False)
df5_data = pd.read_csv('covid.csv',encoding = 'latin1',usecols = ['Country/Region','Recovered'])
top5_poorlyperforming_recovery = df5_data.groupby(["Country/Region"]).max()
top5_poorlyperforming_recovery = top5_poorlyperforming_recovery.sort_values(by = "Recovered",ascending = True)
print ("\n\n>>Task 2-c: The top 5 poorly performing countries, in the order of confirmed, deaths, and recovered cases are: \n%s" % (top5_poorlyperforming_confirmed.head(5)))
print(top5_poorlyperforming_deaths.head(5))
print(top5_poorlyperforming_recovery.head(5))
# 7 points
#Task 2-d: Find out the top 5 poorly performing WHO regions, in the order of confirmed, deaths, and recovered cases.
top5_disasters_confirmed = total_cases_whoregion.groupby(["WHO Region"]).aggregate({'Confirmed':'sum'})
top5_disasters_confirmed = top5_disasters_confirmed.sort_values(by = "Confirmed",ascending = False)
top5_disasters_deaths = total_cases_whoregion.groupby(["WHO Region"]).aggregate({'Deaths':'sum'})
top5_disasters_deaths = top5_disasters_deaths.sort_values(by = "Deaths",ascending = False)
top5_disasters_recovery = total_cases_whoregion.groupby(["WHO Region"]).aggregate({'Recovered':'sum'})
top5_disasters_recovery = top5_disasters_recovery.sort_values(by = "Recovered",ascending = True)
print ("\n\n>>Task 2-d: The top 5 poorly performing WHO regions, in the order of confirmed, deaths, and recovered cases are: \n%s" % (top5_disasters_confirmed.head(5)))
print (top5_disasters_deaths.head(5))
print(top5_disasters_recovery.head(5))

```
